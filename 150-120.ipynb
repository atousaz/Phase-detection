```python
import pandas as pd
import numpy as np
path_120='../120/phase_detection/manual/'
manual_120=pd.read_csv(path_120+'manual.csv',index_col=[0])



path_150='../150/phase_detection/manual/'
manual_150=pd.read_csv(path_150+'manual.csv',index_col=[0])


#manual_150
manual_all=pd.concat([manual_120,manual_150], axis=0)


manual_all
```

```python
manual_all
```

```python
summary=manual_all.groupby(['signal','detector_parameter','parameter','Protected_Phase_Number'])['correct'].count().reset_index().rename(columns={'correct':'count'})
summary.sort_values(['signal','detector_parameter','parameter','count'])
result=summary.loc[summary.groupby(['signal','detector_parameter'])['count'].idxmax()]
result['correct']=np.where(result['parameter']==result['Protected_Phase_Number'],1,0)

g=result.groupby('signal')
detailed_accuracy=pd.DataFrame([])
for sets, data in g:
    accuracy=len(data[data['correct']==1])/len(data)
    df=pd.DataFrame({'signal':sets, 'accuracy':accuracy,'number of detectors':len(data)}, index=[0])
    detailed_accuracy=detailed_accuracy.append(df)
detailed_accuracy
```

```python
result.to_csv('result.csv')
```
